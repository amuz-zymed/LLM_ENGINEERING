# Comparing LLM Providers: Gemini vs Ollama

This notebook demonstrates how to work with different Large Language Model (LLM) providers:

- **Google Gemini**: Cloud-based API from Google (requires API key and internet)
- **Ollama**: Local LLM runner (free, runs on your machine, no API costs)

We'll set up both and compare their usage through the OpenAI-compatible API format.

**Prerequisites:**
- Google API key in `.env` file (for Gemini)
- Ollama installed and running locally (for Ollama)
- Required Python packages installed

Let's explore both approaches!

## Setting up Google Gemini

Google Gemini is a powerful LLM available through Google's API. To use it:

1. Get an API key from [Google AI Studio](https://makersuite.google.com/app/apikey)
2. Add it to a `.env` file as `GOOGLE_API_KEY=your_key_here`
3. Use the OpenAI-compatible endpoint provided by Google

**Pros:** High performance, latest models, cloud-hosted
**Cons:** Requires API key, internet connection, potential costs

The code below checks if your API key is properly configured.

```python
EMINI_BASE_URL = "https://generativelanguage.googleapis.com/v1beta/openai/"

load_dotenv(override=True)

google_api_key = os.getenv("GOOGLE_API_KEY")

if not google_api_key:
    print("No API key was found - please be sure to add your key to the .env file, and save the file! Or you can skip the next 2 cells if you don't want to use Gemini")
elif not google_api_key.startswith("AIz"):
    print("An API key was found, but it doesn't start AIz")
else:
    print("API key found and looks good so far!")

```

```python
gemini = OpenAI(base_url=GEMINI_BASE_URL, api_key=google_api_key)
```

Let's explore both approaches!

```python


response = gemini.chat.completions.create(model="gemini-2.5-flash", messages=[{"role": "user", "content": "Tell me a fun fact"}])

response.choices[0].message.content
```

## Setting up Ollama (Local LLM)

Ollama allows you to run LLMs locally on your machine without API keys or internet requirements.

**Setup steps:**
1. Install Ollama: `curl -fsSL https://ollama.ai/install.sh | sh`
2. Pull a model: `ollama pull llama3.2`
3. Start the model: `ollama run llama3.2`

**Pros:** Free, private, no API limits, works offline
**Cons:** Requires local resources (CPU/GPU), model downloads can be large

The code below creates an OpenAI client that connects to your local Ollama instance.

```python
OLLAMA_BASE_URL = "http://localhost:11434/v1"

ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
```
