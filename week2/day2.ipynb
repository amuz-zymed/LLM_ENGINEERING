# User Interface time

Gradio is an open-source Python library used to quickly build interactive web interfaces for machine learning models or any Python function—without needing front-end development skills.

```python
import gradio as gr
```

```python
def shout(text):
    print(f"Shout has been called with input: {text}")

    return text.upper()
```

```python
shout("dobbyyy is freeeee")
```

Building a UI that launches webserver running locally,
that will serve up a react ui to bring this to forefront to allow user to interact with the abv code to call shout


flagging_mode="never" is used when you want to completely disable logging of user interactions, typically for privacy, security, and controlled environments, especially when dealing with sensitive or potentially exploitable data

```python
# creating a new interface
# passing it the function we want to use, the input component and the output component
# usage of fn is optional
gr.Interface(
    fn=shout, 
    inputs="textbox", 
    outputs="textbox", 
    flagging_mode="never"   # flagging_mode="never" ensures zero data retention
    ).launch()
#callback function that will be called when the user interacts with the interface
```

# share feature


Gradio’s share=True feature enables public access to a locally hosted application by creating a secure HTTP tunnel between the user’s machine and Gradio’s remote servers. This allows external users to interact with the application via a temporary public URL without deploying the application to a cloud environment. However, computation still occurs on the local machine, and the feature introduces potential security and privacy risks.


```python
gr.Interface(fn=shout, inputs="textbox", outputs="textbox", flagging_mode="never").launch(share=True)
```

# Forcing dark mode

**dark mode**

```python
force_dark_mode = """
function refresh() {
    const url = new URL(window.location);
    if (url.searchParams.get('__theme') !== 'dark') {
        url.searchParams.set('__theme', 'dark');
        window.location.href = url.href;
    }
}
"""
gr.Interface(fn=shout, inputs="textbox", outputs="textbox", flagging_mode="never").launch(js=force_dark_mode)
```

**light mode**

```python
force_light_mode = """
function refresh() {
    const url = new URL(window.location);
    if (url.searchParams.get('__theme') !== 'light') {
        url.searchParams.set('__theme', 'light');
        window.location.href = url.href;
    }
}
"""

interface = gr.Interface(
    fn=shout, 
    inputs="textbox", 
    outputs="textbox", 
    flagging_mode="never"
)

interface.launch(js=force_light_mode) # wont work if your system thee is set to dark mode
```

# using stream

```python
def stream_groq(prompt):
    messages = [
        {"role": "system", "content": system_message},
        {"role": "user", "content": prompt}
    ]

    stream = groq.chat.completions.create(
        model="llama-3.3-70b-versatile", 
        messages=messages,
        stream=True
        )
    
    result = ""
    for chunk in stream:
        content = chunk.choices[0].delta.content or ""
        result += content
        yield result
```

```python

message_input = gr.Textbox(label="Your message", info="Enter a message", lines=7)
message_output = gr.Textbox(label="Response", lines=9)


view = gr.Interface(
    fn=stream_groq, 
    title="Groq",
    inputs=[message_input], 
    outputs=[message_output],
    examples=[
        "Explain transformer architecture to an aspiring ML engineer",
        "What are the benefits of meditation?",
    ], 
    flagging_mode="never"
)

view.launch()
```

```python
def stream_gemini(prompt):
    messages = [
        {"role": "system", "content": system_message},
        {"role": "user", "content": prompt}
    ]

    stream = gemini.chat.completions.create(
        model="gemini-3-flash-preview", 
        messages=messages,
        stream=True
        )
    
    result = ""
    for chunk in stream:
        content = chunk.choices[0].delta.content or ""
        result += content
        yield result
```

# practice

```python
def stream_model(prompt, model):
    if model == "groq":
        result = stream_groq(prompt)
    elif model == "gemini":
        result = stream_gemini(prompt)
    else:
        raise ValueError("Model not supported")
    yield from result  # same as  for chunk in result: yeild chunk
```

```python
message_input = gr.Textbox(label="Your message", info="Enter a message", lines=7)
model_selector = gr.Dropdown(["groq", "gemini"], label="Select model", value="GPT")
message_output = gr.Textbox(label="Response", lines=9)

view = gr.Interface(
    fn=stream_model, 
    title="LLM Stream",
    inputs=[message_input, model_selector], 
    outputs=[message_output],
    examples=[
        ["Explain transformer architecture to an aspiring ML engineer", "groq"],
        ["What are the benefits of meditation?", "gemini"],
    ], 
    flagging_mode="never"
)

view.launch()
```

# Building a Company brochure Generator

```python
from bs4 import BeautifulSoup
import requests


# Standard headers to fetch a website
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36"
}


def fetch_website_contents(url):
    """
    Return the title and contents of the website at the given url;
    truncate to 2,000 characters as a sensible limit
    """
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, "html.parser")
    title = soup.title.string if soup.title else "No title found"
    if soup.body:
        for irrelevant in soup.body(["script", "style", "img", "input"]):
            irrelevant.decompose()
        text = soup.body.get_text(separator="\n", strip=True)
    else:
        text = ""
    return (title + "\n\n" + text)[:2_000]


def fetch_website_links(url):
    """
    Return the links on the webiste at the given url
    I realize this is inefficient as we're parsing twice! This is to keep the code in the lab simple.
    Feel free to use a class and optimize it!
    """
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, "html.parser")
    links = [link.get("href") for link in soup.find_all("a")]
    return [link for link in links if link]
```

```python

# Again this is typical Experimental mindset - I'm changing the global variable we used above:

system_message = """
You are an assistant that analyzes the contents of a company website landing page
and creates a short brochure about the company for prospective customers, investors and recruits.
Respond in markdown without code blocks.
"""
```

```python
def stream_brochure(company_name, url, model):
    yield ""
    prompt = f"Please generate a company brochure for {company_name}. Here is their landing page:\n"
    prompt += fetch_website_contents(url)
    if model=="GPT":
        result = stream_gemini(prompt)
    elif model=="Claude":
        result = stream_groq(prompt)
    else:
        raise ValueError("Unknown model")
    yield from result
```

```python
name_input = gr.Textbox(label="Company name:")
url_input = gr.Textbox(label="Landing page URL including http:// or https://")
model_selector = gr.Dropdown(["Gemini", "Groq"], label="Select model", value="Gemini")
message_output = gr.Markdown(label="Response:")

view = gr.Interface(
    fn=stream_brochure,
    title="Brochure Generator",
    inputs=[name_input, url_input, model_selector],
    outputs=[message_output],
    examples=[
            ["Hugging Face", "https://huggingface.co", "Gemini"],
            ["Edward Donner", "https://edwarddonner.com", "Groq"]
        ],
    flagging_mode="never"
    )
view.launch()
```
