```python
system_message = "You are a helpful assistant"
def message_groq(prompt):
    messages = [
        {"role": "system", "content": system_message},
        {"role": "user", "content": prompt}
    ]

    response = groq.chat.completions.create(
        model="llama-3.3-70b-versatile", 
        messages=messages
        )
    return response.choices[0].message.content
```

# using stream

```python
def stream_groq(prompt):
    messages = [
        {"role": "system", "content": system_message},
        {"role": "user", "content": prompt}
    ]

    stream = groq.chat.completions.create(
        model="llama-3.3-70b-versatile", 
        messages=messages,
        stream=True
        )
    
    result = ""
    for chunk in stream:
        content = chunk.choices[0].delta.content or ""
        result += content
        yield result
```

```python

message_input = gr.Textbox(label="Your message", info="Enter a message", lines=7)
message_output = gr.Textbox(label="Response", lines=9)


view = gr.Interface(
    fn=stream_groq, 
    title="Groq",
    inputs=[message_input], 
    outputs=[message_output],
    examples=[
        "Explain transformer architecture to an aspiring ML engineer",
        "What are the benefits of meditation?",
    ], 
    flagging_mode="never"
)

view.launch()
```

```python
def stream_gemini(prompt):
    messages = [
        {"role": "system", "content": system_message},
        {"role": "user", "content": prompt}
    ]

    stream = gemini.chat.completions.create(
        model="gemini-3-flash-preview", 
        messages=messages,
        stream=True
        )
    
    result = ""
    for chunk in stream:
        content = chunk.choices[0].delta.content or ""
        result += content
        yield result
```

# practice

```python
def stream_model(prompt, model):
    if model == "groq":
        result = stream_groq(prompt)
    elif model == "gemini":
        result = stream_gemini(prompt)
    else:
        raise ValueError("Model not supported")
    yield from result  # same as  for chunk in result: yeild chunk
```

```python
message_input = gr.Textbox(label="Your message", info="Enter a message", lines=7)
model_selector = gr.Dropdown(["groq", "gemini"], label="Select model", value="GPT")
message_output = gr.Textbox(label="Response", lines=9)

view = gr.Interface(
    fn=stream_model, 
    title="LLM Stream",
    inputs=[message_input, model_selector], 
    outputs=[message_output],
    examples=[
        ["Explain transformer architecture to an aspiring ML engineer", "groq"],
        ["What are the benefits of meditation?", "gemini"],
    ], 
    flagging_mode="never"
)

view.launch()
```

# Building a Company brochure Generator

```python
from scraper import fetch_website_contents
```

```python

# Again this is typical Experimental mindset - I'm changing the global variable we used above:

system_message = """
You are an assistant that analyzes the contents of a company website landing page
and creates a short brochure about the company for prospective customers, investors and recruits.
Respond in markdown without code blocks.
"""
```

```python
def stream_brochure(company_name, url, model):
    yield ""
    prompt = f"Please generate a company brochure for {company_name}. Here is their landing page:\n"
    prompt += fetch_website_contents(url)
    if model=="GPT":
        result = stream_gemini(prompt)
    elif model=="Claude":
        result = stream_groq(prompt)
    else:
        raise ValueError("Unknown model")
    yield from result
```

```python
name_input = gr.Textbox(label="Company name:")
url_input = gr.Textbox(label="Landing page URL including http:// or https://")
model_selector = gr.Dropdown(["Gemini", "Groq"], label="Select model", value="Gemini")
message_output = gr.Markdown(label="Response:")

view = gr.Interface(
    fn=stream_brochure,
    title="Brochure Generator",
    inputs=[name_input, url_input, model_selector],
    outputs=[message_output],
    examples=[
            ["Hugging Face", "https://huggingface.co", "Gemini"],
            ["Edward Donner", "https://edwarddonner.com", "Groq"]
        ],
    flagging_mode="never"
    )
view.launch()
```
