# Training vs Inference time scaling

tiny model + minimal reasoning => wrong answer

tiny model + low reasoning(dialed up from minimal) => correct answer

dial up the model size + minimal reasoning => correct answer 

* training time -> model size

* inference time -> reasoning

# An adversial conversation between Chatbots


```python
groq_model1 = "llama-3.3-70b-versatile"
groq_model2 = "llama-3.3-70b-versatile"

groq_model1_system = "You are a chatbot who is very argumentative; \
you disagree with anything in the conversation and you challenge everything, in a snarky way."

groq_model2_system = "You are a very polite, courteous chatbot. You try to agree with \
everything the other person says, or find common ground. If the other person is argumentative, \
you try to calm them down and keep chatting."

groq_model1_messages = ["Hi there"]
groq_model2_messages = ["Hi"]
```

```python
def call_groq_model1():
    messages = [{"role": "system", "content": groq_model1_system}]
    for groq1, groq2 in zip(groq_model1_messages, groq_model2_messages):
        messages.append({"role": "assistant", "content": groq1})
        messages.append({"role": "user", "content": groq2})
    response = groq.chat.completions.create(model=groq_model1, messages=messages)
    return response.choices[0].message.content
```

```python
call_groq_model1()
```

```python
def call_groq_model2():
    messages = [{"role": "system", "content": groq_model2_system}]
    for groq1, groq2 in zip(groq_model1_system, groq_model2_system):
        messages.append({"role": "user", "content": groq1})
        messages.append({"role": "assistant", "content": groq2})
    messages.append({"role": "user", "content": groq_model1_system[-1]})
    response = groq.chat.completions.create(model=groq_model2, messages=messages)
    return response.choices[0].message.content
```

```python
call_groq_model2()
```

```python
groq_model1_messages = ["Hi there"]
groq_model2_messages = ["Hi"]

display(Markdown(f"### Groq1:\n{groq_model1_messages[0]}\n"))
display(Markdown(f"### Groq2:\n{groq_model2_messages[0]}\n"))

for i in range(5):
    g1 = call_groq_model1()
    display(Markdown(f"### Groq1:\n{g1}\n"))
    groq_model1_messages.append(g1)
    groq_model2_messages.append(g1)

    g2 = call_groq_model2()
    display(Markdown(f"### Groq2:\n{g2}\n"))
    groq_model2_messages.append(g2)
```
